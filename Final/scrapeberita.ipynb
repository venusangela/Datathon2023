```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import csv
import pandas as pd

hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

```python
df = pd.DataFrame(columns=['Tanggal', 'Bulan', 'Tahun', 'Jam', 'Judul', 'Isi', 'Link'])

for i in range(1,78):
    url = f"https://www.detik.com/search/searchall?query=penyebab+macet&siteid=2&sortby=time&page={i}"
    response = requests.get(url)
    webpage = response.text
    soup = BeautifulSoup(webpage, 'html.parser')

    date = soup.find_all(name='span', class_='date')
    judul_berita = soup.find_all(name='h2', class_="title")
    arc = soup.find_all(name='article')
    link = [arc[j].find("a")['href'] for j in range(len(arc))]
    
    for n in range(len(judul_berita)):
        response1 = requests.get(link[n])
        webpage1 = response1.text
        soup1 = BeautifulSoup(webpage1, 'html.parser')
        body = soup1.find_all(name='div', class_='detail__body itp_bodycontent_wrapper')
        try:
            texts = body[0].find_all('p')
            txt = [t.text for t in texts]
            content = ''.join(txt).replace('\n', '').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')

            tanggal = date[n].text.split(",")[1].split(" ")[1]
            bulan = date[n].text.split(",")[1].split(" ")[2]
            tahun = date[n].text.split(",")[1].split(" ")[3]
            jam = date[n].text.split(",")[1].split(" ")[4]

            new_row = {'Tanggal': tanggal, 
                        "Bulan":bulan,
                        'Tahun':tahun,
                        'Jam': jam, 
                        'Judul': judul_berita[n].text,
                        'Isi': content, 
                        'Link': link[n] 
                        }
            df = df.append(new_row, ignore_index=True)
        except:
            pass
```
