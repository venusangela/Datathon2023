```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
```

# Data Preprocessing

```python
path = "Traffic Condition Dataset V3"
# df = pd.DataFrame(columns={'Days','RushHour','Weather','Temperature','Humidity','Density'})

for i,file in enumerate(os.listdir(path)):
    daerah = file.split(".")[0]
    if i == 0:
        df = pd.read_csv(f"{path}/{file}")
        df['Daerah'] = [daerah for j in range(len(df))]
    else:
        data = pd.read_csv(f"{path}/{file}")
        data['Daerah'] = [daerah for j in range(len(data))]
        df = df.append(data, ignore_index=True)
```

# Feature Engineering

```python
def is_weekend(day):
    if day > 4:
        return 0
    return 1

data['weekend'] = data.apply(lambda row: is_weekend(row['Days']), axis=1)
data
```

```python
fig, axes = plt.subplots(nrows=1,ncols=1, figsize=(25,6))
ax = sns.countplot(ax=axes, data=data, x=data['weekend'], hue=data['Density'])
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
def is_peak(weekend, rushhour):
    if weekend > 0 and rushhour > 0:
        return 0
    return 1

data['Peak'] = data.apply(lambda row: is_peak(row['weekend'], row['RushHour']), axis=1)
data
```

```python
fig, axes = plt.subplots(nrows=1,ncols=1, figsize=(25,6))
ax = sns.countplot(ax=axes, data=data, x=data['Peak'], hue=data['Density'])
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```
