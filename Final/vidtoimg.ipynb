```python
import cv2

# Open the video file
video_path = 'VID_20230815_204054.mp4'
cap = cv2.VideoCapture(video_path)

# Check if the video file was opened successfully
if not cap.isOpened():
    print("Error opening video file")
    exit()

# Initialize variables
frame_count = 0
time_interval = 10  # Time interval in seconds

while True:
    # Calculate the target frame position based on time interval
    target_frame_pos = int(cap.get(cv2.CAP_PROP_FPS) * time_interval * frame_count)

    # Set the video capture object to the target frame position
    cap.set(cv2.CAP_PROP_POS_FRAMES, target_frame_pos)

    # Read the frame from the target position
    ret, frame = cap.read()

    # Break the loop when we've reached the end of the video
    if not ret:
        break

    # Save the frame as an image
    frame_filename = f"frame_mirota/frame_mirota_malam_{frame_count:04d}.jpg"  # You can adjust the filename format
    cv2.imwrite(frame_filename, frame)

    # Print progress
    print(f"Saved {frame_filename} at {target_frame_pos / cap.get(cv2.CAP_PROP_FPS):.2f} seconds")

    # Increment frame count
    frame_count += 1

# Release the video capture object
cap.release()

print("Frames extracted and saved successfully")
```
